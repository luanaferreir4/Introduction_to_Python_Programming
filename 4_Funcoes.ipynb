### Argumentos Padres (Default Arguments)

Essa funo abaixo inclui um _argumento padro_:

````python
import math
def volume_cilindro(altura, raio=5):
    pi = math.pi
    return round(pi * raio**2 * altura, 4)

print(volume_cilindro(10)) # (10, 5)           # VALOR DO RAIO OMITIDO.
print(volume_cilindro(10, 5)) # (10, 5)
print(volume_cilindro(10, 3))
````

Os argumentos padres permitem que as funes usem valores padro quando eles so omitidos como acima.

isso  til pois pode tornar seu cdigo mais conciso nos cenrios que h um valor em comum que voc pode usar para uma varivel, embora voc ainda queira que seja personalizvel.

- Como extrair valores de uma funo passando os argumentos pelo nome:

````python
import math
def volume_cilindro(altura, raio=5):
    pi = math.pi
    return round(pi * raio**2 * altura, 4)

print(volume_cilindro(raio=3, altura=10)) # por nome.
````

- Minha soluo:

````python
# write your function here
def weeks_days(days):
    soma = 0
    semana = 7
    contador = days
    while days >= semana:
        if days == semana:
            return 1, 0
        else:
            while contador >= semana:
                soma += 1
                contador -= semana
            return soma, contador
    return 0, contador

def readable_timedelta(days):
    return '{} week(s) and {} day(s).'.format(weeks_days(days)[0], weeks_days(days)[1])
            
# test your function
print(readable_timedelta(1))
````

````python
def readable_timedelta(dias):
    # use integer division to get the number of weeks
    semanas = dias // 7
    # use % to get the number of days that remain
    resto = dias % 7
    return "{} week(s) and {} day(s).".format(semanas, resto)

readable_timedelta(22)
````

- Algoritmo mais completo feito por mim utilizando a mesma lgica:

````python
def readable_datetime(dias):
    anos = dias // 365
    resto_anos = dias % 365
    meses = resto_anos // 30
    resto_meses = resto_anos % 30
    semanas = resto_meses // 7
    resto = resto_meses % 7
    return {'anos': anos, 'meses': meses, 'semanas': semanas, 'dias': resto}


print(readable_datetime(400))
````

## Variable Scope (Escopo Varivel)
- O escopo se refere a quais partes de um programa uma varivel pode ser referenciada ou usada.<br>
    - Se uma varivel  criada dentro de uma funo, s pode ser usada dentro dessa funo.
- Cosidere essas duas funes:

````python
def word_count(document, search_term):
    words = document.split()
    answer = 0
    for word in words:
        if word == search_term:
            answer += 1
    return answer
````

````python
def nearest_square(limit):
    answer = 0
    while (answer + 1) ** 2 < limit:
        answer += 1
    return answer**2

print(answer)
````

> Essa ltima __deu erro__ pois _chamamos uma varivel que estava dentro de um escopo de funo no escopo global_, onde no existe uma varivel com esse nome, no podemos chamar variveis de funes em escopos globais atravs de print ou afins. Podemos apenas usar variveis de escopos de funes dentro de funes.

````python
ans = 10

def show_ans():
    print(ans)
    
show_ans() 
````

__Importante:__ j o contrrio podemos fazer, usar variveis de escopo global dentro de escopos de funes.

````python
# This works fine
word = "hello"

def some_function():
    print(word)

some_function()
````

Observe que ainda podemos acessar o valor da varivel global `word` dentro desta funo. Porm, o valor de uma varivel global no pode ser modificado dentro da funo. Se voc quiser modificar o valor dessa varivel dentro dessa funo, ela deve ser passada como um argumento. Voc ver mais sobre isso no prximo teste.

- __Mais sobre escopo varivel__

Ao programar, muitas vezes voc descobrir que ideias semelhantes surgem repetidamente. Voc usar variveis para coisas como contagem, iterao e acmulo de valores a serem retornados. Para escrever um cdigo legvel, voc desejar usar nomes semelhantes para ideias semelhantes. Assim que voc juntar vrias partes do cdigo (por exemplo, vrias funes ou chamadas de funo em um nico script), poder descobrir que deseja usar o mesmo nome para dois conceitos separados.

Felizmente, voc no precisa criar novos nomes indefinidamente. A reutilizao de nomes para objetos  aceitvel, desde que voc os mantenha em escopos separados.

___Boa prtica:___  melhor definir as variveis no menor escopo em que sero necessrias. Embora as funes possam se referir a variveis definidas em um escopo maior, isso raramente  uma boa ideia, pois voc pode no saber quais variveis voc definiu se o seu programa tem muitas variveis.

- __Soluo do questionrio: escopo varivel__

O trecho de cdigo na pgina anterior  repetido aqui:

````python
egg_count = 0

def buy_eggs():
    egg_count += 12 # purchase a dozen eggs

buy_eggs()
````

Esse cdigo causa um `UnboundLocalError`, porque a varivel `egg_count` na primeira linha tem escopo global. Observe que ele no  passado como um argumento para a funo, ento a funo assume que a `egg_count` referncia  a varivel global.

No ltimo vdeo, voc viu que dentro de uma funo podemos imprimir o valor de uma varivel global com sucesso e sem erros. Isso funcionou porque estvamos simplesmente acessando o valor da varivel. Se tentarmos alterar ou reatribuir essa varivel global, no entanto, como fazemos neste cdigo, obteremos um erro. Python no permite funes para modificar variveis que no esto no escopo da funo.

Uma maneira melhor de escrever isso seria:

````python
egg_count = 0

def buy_eggs(count):
    return count + 12  # purchase a dozen eggs

egg_count = buy_eggs(egg_count)
print(egg_count)
````

### Documentao

- Documentao  usada para fazer seu cdigo mais fcil de entender e usar. _Funes so especialmente de fcil compreenso pois eles geralmente usam strings de documentao, ou docstrings_. __Docstrings__ so um tipo de comentrio usado para explicar o propsito de uma funo, e como ela deveria ser usada. Aqui h uma funo para a funo da densidade populacional com uma docstring:


- Se a funo for simples:

````python
def densidade_populacional(populacao, area):
    """Calcula a densidade populacional de uma rea."""
    return populacao / area
````

- Se a funo for mais complexa:

````python
def densidade_populacional(populacao, area):
    """Calcula a densidade populacional de uma rea.
    Args:
        populacao: int. a populao da rea.
        area: int ou float. Essa funo  unit-agnostic (no depende de unidades
        especficas para funcionar ou ser avaliado), se voc passar em valores nos termos
        de km ao quadrado ou milhas ao quadrado a funo retornar a densidade nessas unidades.
    
    Explaining more:
        "Unit-agnostic"  uma expresso usada para descrever algo que no depende de unidades
        especficas para funcionar ou ser avaliado. Em outras palavras, "unit-agnostic"
        significa que a medida ou valor no est vinculado a uma unidade especfica de medida,
        como metros, ps, quilogramas ou libras.
    
    Returns
    populacao / area. A densidade populacional de uma rea.
    """
    return populacao / area
````

## Expresses Lambda

- Para criarmos funes annimas podemos usar expresses lambda em Python. Ou seja, funes sem nome.
-  muito til para criar funes rpidas que no sero muito necessrias mais pra frente no programa.<br>

_Depois vamos aprender higher-order functions._<br>
- __O que so higher-order functions?__ 

Higher-order functions so funes em programao que podem aceitar outras funes como argumentos ou retornar outras funes como resultado. Em outras palavras, higher-order functions so funes que operam em outras funes.

Essas funes so usadas para aumentar a reutilizao de cdigo, a modularidade e a legibilidade do cdigo, pois permitem que funes sejam passadas como argumentos para outras funes, tornando-as mais flexveis e reutilizveis. Alm disso, higher-order functions tambm permitem a criao de funes genricas que podem ser usadas para manipular diferentes tipos de dados.

- Vamos comparar uma funo comum com uma expresso lambda:

````python
def multiply(x):
    return x * 2
````

````python
multiply = lambda x: x * 2  # equivalente em uma funo lambda
multiply(2)
````

````python
multiply = lambda x, y: x * y
multiply(5, 2)
````

- Expresses lambda __no so ideais para funes complexas mas para funes simples e curtas__.

- __Quiz: Lambda com Map__

`map()`  uma funo built-in higher-order que usa uma funo e iterveis como inputs, e retorna um iterador que aplica a funo a cada elemento do itervel. O cdigo abaixo usa `map()` para encontrar a mdia de cada lista em nmeros para criar mdias de listas. Execute para ver o que acontece.

Reescreva este cdigo para ser mais conciso, substituindo a funo `mean` por uma expresso lambda definida dentro da chamada a `map()`.

````python
numbers = [
              [34, 63, 88, 71, 29],
              [90, 78, 51, 27, 45],
              [63, 37, 85, 46, 22],
              [51, 22, 34, 11, 18]
           ]

def mean(num_list):
    return sum(num_list) / len(num_list)

averages = list(map(mean, numbers))
print(averages)
````

````python
numbers = [
              [34, 63, 88, 71, 29],
              [90, 78, 51, 27, 45],
              [63, 37, 85, 46, 22],
              [51, 22, 34, 11, 18]
           ]

mean = lambda num_list: sum(num_list) / len(num_list)
averages = list(map(mean, numbers))
print(averages)
````

- __Quiz: Lambda com Filter__

`filter()`  uma funo built-in higher-order que recebe uma funo e iterveis como inputs e retorna um iterador com os elementos do itervel que a funo retorna True. O cdigo abaixo usa `filter()`  para pegar os nomes nas cidades que tem menos de 10 caracteres para criar a lista `short_cities`. Execute para ver o que acontece.

Reescreva esse cdigo para ser mais conciso, substituindo a funo `is_short` por uma expresso lambda definida dentro da chamada a `filter()`.

````python
cities = ["New York City", "Los Angeles", "Chicago", "Mountain View", "Denver", "Boston"]

def is_short(name):
    return len(name) < 10

short_cities = list(filter(is_short, cities))
print(short_cities)
````

````python
cities = ["New York City", "Los Angeles", "Chicago", "Mountain View", "Denver", "Boston"]

is_short = lambda city: len(city) < 10
short_cities = list(filter(is_short, cities))
print(short_cities)
````
