## Conditional Statements
- _Tradução_: Estruturas Condicionais

Muitas vezes queremos executar um código apenas em alguma condição em particular. <br>
Para isso usamos as _Estruturas Condicionais_ que se parecem com:

```python
balanco_cel = float(input('Créditos restantes no celular:'))
balanco_banco = float(input('Saldo do banco:'))

print(balanco_cel, balanco_banco)

if balanco_cel < 5:
    balanco_cel += 10
    balanco_banco -= 10
    
print(balanco_cel, balanco_banco)
```

### Estrutura `if`
Esse foi um exemplo da _estrutura __`if`___ que executa ou pula o código baseado em se a condição for verdadeira ou falsa. (True or False)<br>

- Se a condição for `True` ele executa o campo identado.
- Do contrário, se for `False` ele apenas pula se não tiver `else`.
### Estrutura `else`
- Se tiver `else`, ele executa se o resultado nem de `if`'s e nem de `elif`'s forem igual a True. `else` nesse caso é a última opção.

```python
n = 145

if n % 2 == 0:
    print('Número ' + str(n) + ' é par.')  # True
else:
    print('Número ' + str(n) + ' é ímpar.')  # False
```

### Estrutura `elif`
Se você tiver mais que 2 possibilidades, você pode usar `elif` para lidar com a situação.

```python
estacao = 'inverno'

if estacao == 'primavera':
    print('plante o jardim!')
elif estacao == 'verão':
    print('molhe o jardim!')
elif estacao == 'outono':
    print('colha o que plantou!')
elif estacao == 'inverno':
    print('fique dentro de casa!')
else:
    print('estação não reconhecida')
```

__Prática: Qual prêmio?__<br>
Escreva uma instrução if que permita a um competidor saber quais desses prêmios ele ganhou com base no número de pontos que marcou, que é armazenado na variável inteira `pontos`.

> - Pontos - Prêmio
> - 1 - 50 = coelho de madeira
> - 51 - 150 = sem prêmio
> - 151 - 180 = hortelã fina
> - 181 - 200 = pinguim<br>
Todos os limites inferior e superior aqui são inclusivos e os pontos só podem assumir valores inteiros positivos até 200.

Em sua estrutura `if`, atribua a variavel `resultado` a uma string segurando uma mensagem apropriada baseada no valor de `pontos`. Se ganharam um prêmio, a mensagem deve estar como`"Parabéns! Você ganhou um(a) [nome do prêmio]!"`. Se não há prêmio, a mensagem deve estar como `"Oh querido, sem prêmios dessa vez."`

```python
points = 174  # use this input to make your submission

# write your if statement here
if points <= 50:
    result = "Congratulations! You won a wooden rabbit!"
elif points <= 150:
    result = "Oh dear, no prize this time."
elif points <= 180:
    result = "Congratulations! You won a wafer-thin mint!"
else:
    result = "Congratulations! You won a penguin!"


print(result)
```

Nós usamos `<=`  ao invés de `<`, pois os valores mínimos e máximos para considerar ou não os prêmios são inclusivos. Note que em cada condição, nós checamos se os pontos estão na faixa de premiação verificando se os pontos são menores ou iguais ao limite superior; não precisávamos verificar se era maior que o limite abaixo. Vamos ver porque este é o caso:
- Quando `points = 174`, é primeiro avaliado se `points <= 50`, que é igual a `False`. Não temos que checar se também é maior que 0. Desde que isso é declarado no problema que `points` sempre será um inteiro positivo até 200.
- Como a primeira condição é avaliada como falsa, ele passa para verificar a próxima condição, `points <= 150`. __Não precisamos verificar se também é maior que 50 aqui!__ Já sabemos que este é o caso porque a primeira condição deve ser avaliada como False para chegar a este ponto. Se sabemos que `points <= 50` é False, então `points > 50` deve ser True!
- Por fim, verificamos se `points <= 180`, que resulta em True. Agora sabemos que os pontos estão na faixa de 151 a 180.
- A última faixa de premiação, 181-200, é apanhada na cláusula else, pois não há outro valor possível do prêmio depois de verificadas as condições anteriores.

- __Quiz: Adivinhe Meu Número__<br>
Você decide que quer jogar um jogo em que está escondendo um número de alguém. Armazene esse número em uma variável chamada 'resposta'. Outro usuário fornece um número chamado 'palpite'. Ao comparar o palpite com a resposta, você informa ao usuário se o palpite é muito alto ou muito baixo.<br>
Preencha os condicionais abaixo para informar o usuário sobre como seu palpite se compara à resposta.

```python
# '''
# You decide you want to play a game where you are hiding 
# a number from someone.  Store this number in a variable 
# called 'answer'.  Another user provides a number called
# 'guess'.  By comparing guess to answer, you inform the user
# if their guess is too high or too low.

# Fill in the conditionals below to inform the user about how
# their guess compares to the answer.
# '''
resposta = 3
palpite = 3

if palpite < resposta:
    resultado = "Ops! Seu palpite foi baixo demais."
elif palpite > resposta:
    resultado = "Ops! Seu palpie foi alto demais."
elif palpite == resposta:
    resultado = "Legal! Seu palpite correspondeu a resposta!"

print(resultado)
```

- __Questionário: Imposto de Compra__<br>
Dependendo de onde um indivíduo é, precisamos tributá-lo adequadamente. Os estados de CA, MN e NY têm impostos de 7,5%, 9,5% e 8,9%, respectivamente. Use esta informação para obter o valor de uma compra e o estado correspondente para garantir que sejam tributados pelo valor correto.

```python
state = "CA"
purchase_amount = 20 # dado aleatório, é o dinheiro de compra sem o imposto.

if state == "CA":
    tax_amount = .075
    total_cost = purchase_amount*(1+tax_amount)
    result = "Since you're from {}, your total cost is {}.".format(state, total_cost)

elif state == "MN":
    tax_amount = .095
    total_cost = purchase_amount*(1+tax_amount)
    result = "Since you're from {}, your total cost is {}.".format(state, total_cost)

elif state == "NY":
    tax_amount = .089
    total_cost = purchase_amount*(1+tax_amount)
    result = "Since you're from {}, your total cost is {}.".format(state, total_cost)

print(result)
```

### Expressões Booleanas Complexas
Em `if` algumas vezes as instruções usamos expressões booleanas mais complicadas para suas condições. Eles podem conter vários operadores de comparação, operadores lógicos e até cálculos. Como exemplo:

Para condições realmente complexas, você pode precisar combinar alguns `and`'s, `or`'s e `not`'s juntos. Use parênteses se precisar deixar as combinações mais claras.

Por mais simples ou complexa que seja, a condição em uma estrutura if deve ser uma expressão booleana avaliada como True ou False e é esse valor que decide se o bloco recuado em uma estrutura if será executado ou não.

#### Bons e Maus Exemplos
Aqui estão algumas coisas que você deve ter em mente ao escrever expressões booleanas para suas estruturas if:

**1. Não use True ou False como condições:** 

```python
# Bad example
if True:
   print('Esse código indentado sempre vai executar.') 
```

Embora `True` seja uma expressão booleana válida, não é útil como condição, pois sempre é avaliada como True, portanto, o código recuado sempre será executado. Da mesma forma, `if False` não é uma condição que você deve usar - a instrução if False nunca seria executada.

```python
esta_frio = False

# Another bad example
if esta_frio or not esta_frio:
    print("Esse código indentado sempre vai executar.")
```

Da mesma forma, é inútil usar qualquer condição que você sabe que sempre será avaliada como True, como no exemplo acima. Uma variável booleana só pode ser True ou False, portanto, `is_cold` ou `not is_cold`é sempre True, e o código recuado sempre será executado.

**2. Tenha cuidado ao escrever expressões que usam operadores lógicos**<br>
Operadores lógicos and, or e not têm significados específicos que não são exatamente iguais aos seus significados simples em inglês. Certifique-se de que suas expressões booleanas estão sendo avaliadas da maneira que você espera.

```python
clima = "chuva"
# Bad example
if clima == "neve" or "chuva": # False or Truthy
    print("Use botas!")
```

Este código é válido em Python, mas não é uma expressão booleana, embora pareça uma. A razão é que a expressão à direita do oroperador, "chuva", não é uma expressão booleana - é uma string! Posteriormente, discutiremos o que acontece quando você usa objetos do tipo não booleano no lugar de booleanos.

**3. Não compare uma variável booleana com `== True` ou `== False`**<br>
Essa comparação não é necessária, pois a própria variável booleana é uma expressão booleana.

Essa é uma condição válida, mas podemos tornar o código mais legível usando a própria variável como condição, conforme abaixo.

Se você deseja verificar se um booleano é falso, pode usar o operador `not`.

#### Teste de Valor de Verdade
Se usarmos um objeto não booleano como condição em uma estrutura if no lugar da expressão booleana, o Python verificará seu valor verdadeiro e o usará para decidir se deve ou não executar o código recuado. Por padrão, o valor verdadeiro de um objeto em Python é considerado True, a menos que seja especificado como False na documentação.
Aqui estão a maioria dos objetos internos que são considerados falsos em Python:
- Constantes definidas como falsas: `None`e `False`
- Zero de qualquer tipo numérico: `0`, `0.0`, `0j`, `Decimal(0)`, `Fraction(0, 1)`
- Sequências e coleções vazias: `'""`, `()`, `[]`, `{}`, `set()`, `range(0)`<br><br>
_Exemplo:_

```python
erros = 3
if erros:
    print("Você tem {} erros para consertar!".format(erros))
else:
    print("Nenhum erro para consertar!")
```

Neste código, `erros` tem o valor verdade True porque é um número diferente de zero, então a mensagem de erro é impressa. Esta é uma maneira agradável e sucinta de escrever uma estrutura if .

Você usará uma nova variável `premio` para armazenar um nome de prêmio, caso tenha ganho, e então usará o valor real dessa variável para compor a mensagem de `resultado`. Isso envolverá duas instruções if.

__1ª estrutura condicional:__ atualize o prize para o nome correto do prêmio com base nos pontos.<br>
__2ª estrutura condicional:__ defina o resultado para a frase correta baseado em se a frase correta é avaliada como True ou False.

- Se o prize é None, result deveria estar definido como `"Oh dear, no prize this time."`
- Se prize contem um nome de prêmio, deveria estar definido como `"Congratulations! You won a {}!".format(prize)`. Isso evitará atribuições de resultados múltiplos para prêmios diferentes.<br><br>

```python
points = 174

if points <= 50:
    prize = "wooden rabbit"
elif points <= 150:
    prize = None
elif points <= 180:
    prize = "wafer-thin mint"
else:
    prize = "penguin"
    

if prize:
    result = "Congratulations! You won a {}!".format(prize)
else: 
    result = "Oh dear, no prize this time."
    
print(result)
```

```python
# Melhor maneira:

points = 174

prize = None

if points <= 50:
    prize = "wooden rabbit"
elif 151 <= points <= 180: 
    prize = "wafer-thin mint"
else:
    prize = "penguin"
    
if prize:
    result = "Congratulations! You won a {}!".format(prize)
else: 
    result = "Oh dear, no prize this time."
    
print(result)
```

Primeiro, definimos `prize` como `None` e, em seguida, o atualizamos apenas se cair em uma faixa que resulte em ganhar um prêmio. Isso é feito na primeira instrução if. Em seguida, usamos o valor verdadeiro do prêmio para atribuir o resultado a uma mensagem com base em se um prêmio foi ganho.

## Loops (Estruturas de Repetição)
- _Tradução_: Estruturas de Repetição

Python tem duas formas de loops - loops `for` e loops `while`. Um loop for é usado para "iterar", ou fazer algo repetidamente sobre um iterável.

Um __iterável__ é um objeto que pode retornar cada um de seus elementos por vez. Isso pode incluir tipos de sequência, como strings, listas e tuplas, bem como tipos não sequenciais, como dicionários e arquivos.

__Exemplo__<br>
Vamos quebrar os componentes de um loop `for` em vários, usando este exemplo com a lista cities:

```python
cidades = ['new york city', 'mountain view', 'chicago', 'los angeles']

for cidade in cidades:
    print(cidade.title())
print('Fim!')
```

Os loops nos permitem a repetição de blocos de código, Python tem 2 tipos de loops (`for` e `while`).
- Primeiro vamos aprender sobre `for` loop que podemos iterar em um iterável a partir dele.

- __Para _criar_ uma lista__:

```python
capitalized_cidades = []

for cidade in cidades:
    capitalized_cidades.append(cidade.title())
    
print(capitalized_cidades)
```

- Modificar uma lista é um pouco mais complexo e requer o uso de uma nova função __range__: `range(start, stop, step)`
- `step` é a diferença entre os números de uma sequência.

```python
for i in range(0, 10, 2):
    print(i)
```

```python
print(list(range(1, 10, 2)))
```

```python
print(set(range(0, 4)))
```

```python
cidades = ['new york city', 'mountain view', 'chicago', 'los angeles']
for i in range(len(cidades)):
    cidades[i] = cidades[i].title()
    print(cidades[i])
```

```python
meu_dicionario = {'Khan': 4, 'Ali': 2, 'Luna': 6, 'Mark': 11, 'Pooja': 8, 'Sara': 1}

for chave, valor in meu_dicionario.items():
    meu_dicionario['Ali'] = 10
    
print(meu_dicionario)
```

```python
my_dict = {'nome': 'Luana', 'idade': 20, 'estado civil': 'solteiro'}

for chave, valor in my_dict.items():
    if chave == 'nome':
        my_dict[chave] = 'Leandro'

print(my_dict)
```
